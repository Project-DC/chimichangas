```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randint(low=0, high=1200, size=1000)
Y = np.random.randint(low=0, high=800,  size=1000)
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(X, Y)
```

```python
from sklearn.cluster import DBSCAN
```

```python
data = np.array([[X[i], Y[i]] for i in range(len(X))])
```
